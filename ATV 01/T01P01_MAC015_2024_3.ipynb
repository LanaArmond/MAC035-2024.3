Desenvolva uma rotina computacional para determinar a intensidade e a direção de n forças coplanares concorrentes em um nó.


## Relatório da Atividade

Nosso código foi projetado para resolver problemas bidimensionais relacionados ao cálculo de forças resultantes e à visualização de diagramas de corpo livre. Ele permite calcular a magnitude e a direção da força resultante a partir de uma lista de forças aplicadas a um corpo, bem como representar graficamente as forças individuais e sua resultante. É possível calcular a força resultante utilizando o ângulo em relação ao eixo x e a magnitude das forças do sistema, e também utilizando relações trigonométricas de triângulos quando as informações fornecidas são medidas do eixo X e do eixo Y.

O código é composto por três principais funções:

`calculate_resultant`: Calcula a força resultante e sua direção.

`calculate_force_by_triangle`: Determina o ângulo de uma força a partir de duas medidas (x e y).

`plot_diagram`: Plota um diagrama de corpo livre com as forças individuais e a resultante.

## Bibliotecas Utilizadas

A biblioteca math é utilizada para cálculos trigonométricos e de conversão entre radianos e graus.

A biblioteca matplotlib é usada para criar os diagramas de corpo livre, aproveitando-se da função quiver para desenhar vetores.

## Representação das Forças

As forças são representadas como tuplas no formato (magnitude, ângulo), onde:

- Magnitude: Intensidade da força (em Newtons).

- Ângulo: Direção da força em graus, medidos no sentido anti-horário a partir do eixo X positivo.

## Visualização

Os vetores das forças individuais e da resultante são desenhados no plano cartesiano:

- Cores: Azul para forças individuais e vermelho para a resultante.

- Escala: As forças são escaladas para manter proporções fáceis de interpretar.

- Anotações: Cada vetor é rotulado com sua magnitude e ângulo.

```python
# Função que calcula a força resultante e sua direção com base em uma lista de forças
def calculate_resultant(forces):
    sum_fx = 0  # Componente x da resultante (FRx)
    sum_fy = 0  # Componente y da resultante (FRy)

    # Decompõe cada força do sistema em suas componentes X e Y, somando para FRy e FRx
    for magnitude, angle in forces:
        angle_rad = math.radians(angle)            # Conversão graus -> radianos
        sum_fx += magnitude * math.cos(angle_rad)  # FRx
        sum_fy += magnitude * math.sin(angle_rad)  # FRy

    # Calcula magnitude da força resultante
    resultant = math.sqrt(sum_fx**2 + sum_fy**2)
    # Calcula direção da foça resultante
    direction = math.degrees(math.atan2(sum_fy, sum_fx))

    # Retorna magnitude de FR, direção de FR, FRx e FRy
    return resultant, direction, sum_fx, sum_fy

# Função que calcula a intensidade e o ângulo da força a partir de medidas no plano x e y
def calculate_force_by_triangle(magnitude, x, y):
    angle = math.degrees(math.atan2(y, x))         # Calcula o ângulo em graus
    return magnitude, angle                        # Retorna magnitude e ânguylo em graus

# Função para a plotagem de um diagrama de corpo livre indicando as forças individuais e a força resultante
def plot_diagram(forces, resultant, direction):
    plt.figure(figsize=(8, 8))
    ax = plt.gca()

    # Plotar as forças individuais
    for magnitude, angle in forces:
        angle_rad = math.radians(angle)
        fx = magnitude * math.cos(angle_rad)
        fy = magnitude * math.sin(angle_rad)
        plt.quiver(0, 0, fx, fy, angles='xy', scale_units='xy', scale=1, color='blue', label=f"Força: {magnitude:.2f} N, {angle:.2f}°")

    # Plotar a força resultante
    resultant_fx = resultant * math.cos(math.radians(direction))
    resultant_fy = resultant * math.sin(math.radians(direction))
    plt.quiver(0, 0, resultant_fx, resultant_fy, angles='xy', scale_units='xy', scale=1, color='red', label=f"Resultante: {resultant:.2f} N, {direction:.2f}°")

    # Ajustes no gráfico
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.xlim(-1.5 * resultant, 1.5 * resultant)
    plt.ylim(-1.5 * resultant, 1.5 * resultant)
    plt.xlabel("Eixo X (N)")
    plt.ylabel("Eixo Y (N)")
    plt.title("Diagrama de Corpo Livre")
    plt.legend()
    plt.show()
```

# Problema 01

### Resolução Manual

Para descobrir o ângulo da direção de cada força, utilizamos relações trigonométricas do triângulo retângulo.

- Ângulo relacionado à força de $800N (F1)$:
$α = arctg |\frac{600}{800}| ≈ 36.87^{\circ \:}$

> $F1_x = 800cos(36.87^{\circ \:}) ≈ 640N$

> $F1_y = 800sen(36.87^{\circ \:}) ≈ 480N$

- Ângulo relacionado à força de $424N (F2)$:
$θ = arctg |\frac{900}{560}| ≈ 58.11^{\circ \:}$

> $F2_x = -424cos(58.11^{\circ \:}) ≈ -224N$

> $F2_y = -424sen(58.11^{\circ \:}) ≈ -360$

- Ângulo relacionado à força de $408N (F3)$:
$β = arctg |\frac{900}{480}| ≈ 61.93^{\circ \:}$

> $F3_x = 408cos(61.93^{\circ \:}) ≈ 192N$

> $F3_y = -408sen(61.93^{\circ \:}) ≈ -360N$

Somatório para eixo X e eixo Y:

$FR_x = 640 - 224 + 192 = 608N$

$FR_y = 480 - 360 - 360 = -240N$

Força Resultante:

$FR = \sqrt{FR_x^2 + FR_y^2} ≈ 653.65N$

Direção da força resultante:

$arctg|\frac{FR_y}{FR_x}| = arctg|\frac{-240}{608}| = 21.54^{\circ \:}$

## Resolução pelo Programa
___

```python
# Componentes das forças conforme o problema
forces_data = [
    (800, 800, 600),  # Força de 800 N, x = 800 mm, y = 600 mm
    (424, -560, -900),  # Força de 424 N, x = -560 mm, y = -900 mm
    (408, 480, -900)  # Força de 408 N, x = 480 mm, y = -900 mm
]

# Converter componentes e calcular ângulos
forces = []
for magnitude, x, y in forces_data:
    x_meters = x / 1000  # Converter de mm para m
    y_meters = y / 1000  # Converter de mm para m
    magnitude, angle = calculate_force_by_triangle(magnitude, x_meters, y_meters)
    forces.append((magnitude, angle))

# Calcular a força resultante
resultant, direction, sum_fx, sum_fy = calculate_resultant(forces)

# Exibir os resultados
print(f"=== Resultado ===")
print(f"Intensidade da força resultante: {resultant:.2f} N")
print(f"Direção da força resultante: {direction:.2f}°")
print(f"Componentes da força resultante: Fx = {sum_fx:.2f} N, Fy = {sum_fy:.2f} N")

# Plotar o diagrama
plot_diagram(forces, resultant, direction)
```

Com isso, podemos ver que os resultados converge com os do programa, porém com representação dos ângulos opostos nos quadrantes onde X é negativo.

Os ângulos nos gradrantes onde x é negativo são:

- $θ - 180^{\circ \:} ⟶ 58.11^{\circ \:} - 180^{\circ \:} = -121.89^{\circ \:}$
- $-61.93^{\circ \:}$ é oposto de $β  = 61.09^{\circ \:}$ no círculo trigonométrico

> O ângulo $61.93^{\circ \:}$ está no primeiro quadrante, com uma rotação positiva a partir da linha horizontal (eixo 𝑥).

> O ângulo $-61.93^{\circ \:}$ está no quarto quadrante, com uma rotação negativa (sentido horário) a partir da linha horizontal.

- $-21.54^{\circ \:}$ é oposto de $21.54^{\circ \:}$ no círculo trigonométrico

Somatório para eixo X e eixo Y:

$FR_x = -123 + 41 + 61.3 = -20.7N$

Força Resultante:

## Resolução pelo Programa
___

```python
# Dados do problema da imagem
forces = [
    (150, 180 - 35),  # 150 N a 35° acima do eixo negativo de x (convertido para 145° no sistema trigonométrico)
    (120, 30 + 40),   # 120 N a 70° acima do eixo y positivo (convertido para 60°)
    (80, 40)          # 80 N a 40° acima do eixo x positivo
]

# Cálculo da resultante e direção
resultant, direction, sum_fx, sum_fy = calculate_resultant(forces)

# Exibir resultados
print("=== Resultado ===")
print(f"Intensidade da força resultante: {resultant:.2f} N")
print(f"Direção da força resultante: {direction:.2f}°")
print(f"Componente x da força resultante: {sum_fx:.2f} N")
print(f"Componente y da força resultante: {sum_fy:.2f} N")

# Plotar o diagrama de corpo livre
plot_diagram(forces, resultant, direction)
```
